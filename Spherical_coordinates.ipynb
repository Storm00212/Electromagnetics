```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Create spherical coordinate grid
r = 1  # radius
theta = np.linspace(0, 2 * np.pi, 1000)
phi = np.linspace(0, np.pi, 1000)

theta, phi = np.meshgrid(theta, phi)
```

```python
# Convert to Cartesian coordinates
x = r * np.sin(phi) * np.cos(theta)
y = r * np.sin(phi) * np.sin(theta)
z = r * np.cos(phi)
```

```python
# Plotting
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z, cmap='viridis', alpha=0.7)

ax.set_title("3D Spherical Coordinate System")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

# simulating radius

```python
import numpy as np
import matplotlib.pyplot as plt

r = 1
theta = np.deg2rad(40)
phi = np.deg2rad(50)

x = r * np.sin(phi) * np.cos(theta)
y = r * np.sin(phi) * np.sin(theta)
z = r * np.cos(phi)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection='3d')

# Origin
ax.scatter(0, 0, 0, color='black')

# Radius vector
ax.plot([0, x], [0, y], [0, z], color='red', linewidth=2, label="r")

# Point
ax.scatter(x, y, z, color='blue', s=50)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend()

ax.set_title("Spherical Coordinate Vector Representation")
plt.show()
```

# simulating a specific point

```python
import numpy as np
import matplotlib.pyplot as plt

# Spherical coordinates
r = 1.5
theta = np.deg2rad(45)
phi = np.deg2rad(60)
#
# Convert to Cartesian
x = r * np.sin(phi) * np.cos(theta)
y = r * np.sin(phi) * np.sin(theta)
z = r * np.cos(phi)

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, color='red', s=100)
ax.text(x, y, z, f"({x:.2f}, {y:.2f}, {z:.2f})")

ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([-2, 2])
# setting labels
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

ax.set_title("Point in Spherical Coordinates")

plt.show()
```
