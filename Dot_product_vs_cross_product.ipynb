```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define vectors
A = np.array([3, 0, 0])
B = np.array([2, 2, 0])
```

```python
# Dot product projection
proj_length = np.dot(A, B) / np.linalg.norm(B)
proj_vector = proj_length * (B / np.linalg.norm(B))
```
