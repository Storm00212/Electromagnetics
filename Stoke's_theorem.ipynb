```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create grid
x = np.linspace(-2, 2, 15)
y = np.linspace(-2, 2, 15)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)

# Vector field F = (-y, x, 0)
U = -Y
V = X
W = np.zeros_like(X)

# Curl of F = (0,0,2)
curl_x = np.zeros_like(X)
curl_y = np.zeros_like(X)
curl_z = 2 * np.ones_like(X)

# Create figure
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

# Plot vector field
ax.quiver(X, Y, Z, U, V, W, length=0.2, normalize=True)

# Plot curl vectors (in red)
ax.quiver(X, Y, Z, curl_x, curl_y, curl_z,
          length=0.3, normalize=True)

# Plot surface (unit disk)
theta = np.linspace(0, 2*np.pi, 50)
r = np.linspace(0, 1, 50)
R, T = np.meshgrid(r, theta)
Xs = R * np.cos(T)
Ys = R * np.sin(T)
Zs = np.zeros_like(Xs)

ax.plot_surface(Xs, Ys, Zs, alpha=0.3)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("Visualization of Stokes' Theorem")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Grid
x = np.linspace(-2, 2, 15)
y = np.linspace(-2, 2, 15)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)

# Vector field F = (-y, x, 0)
U = -Y
V = X
W = np.zeros_like(X)

# Curl = (0,0,2)
curl_x = np.zeros_like(X)
curl_y = np.zeros_like(X)
curl_z = 2 * np.ones_like(X)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(X, Y, Z, U, V, W, length=0.2, normalize=True)
ax.quiver(X, Y, Z, curl_x, curl_y, curl_z, length=0.3, normalize=True)

# Surface (unit disk)
theta = np.linspace(0, 2*np.pi, 40)
r = np.linspace(0, 1, 40)
R, T = np.meshgrid(r, theta)
Xs = R * np.cos(T)
Ys = R * np.sin(T)
Zs = np.zeros_like(Xs)

ax.plot_surface(Xs, Ys, Zs, alpha=0.3)

ax.set_title("Stokes Visualization - Example 1")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Grid
x = np.linspace(-6, 6, 25)
y = np.linspace(-6, 6, 25)
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)

# Convert to polar
rho = np.sqrt(X**2 + Y**2)
phi = np.arctan2(Y, X)

# Field components in Cartesian
A_rho = rho * np.cos(phi)
A_phi = np.sin(phi)

Ax = A_rho * np.cos(phi) - A_phi * np.sin(phi)
Ay = A_rho * np.sin(phi) + A_phi * np.cos(phi)
Az = np.zeros_like(X)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(X, Y, Z, Ax, Ay, Az, length=0.5, normalize=True)

# Sector surface (rho 25, phi 3060)
phi_s = np.linspace(np.pi/6, np.pi/3, 40)
rho_s = np.linspace(2, 5, 40)
R, P = np.meshgrid(rho_s, phi_s)

Xs = R * np.cos(P)
Ys = R * np.sin(P)
Zs = np.zeros_like(Xs)

ax.plot_surface(Xs, Ys, Zs, alpha=0.3)

ax.set_title("Stokes Visualization - Cylindrical Sector")
plt.show()
```
